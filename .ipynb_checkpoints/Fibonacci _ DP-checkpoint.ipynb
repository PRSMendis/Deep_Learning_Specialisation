```python
def fib(n):
    arr = [0,1,1]
    try:
        if arr[n]:
            return arr[n]
    except IndexError:
        result = fib(n-1) + fib(n-2)
        arr[n] = result
        return result
```

```python
import numpy as np
def fib(n):
    arr = np.array([0,1,1])
    if 0 <= n < len(arr):
        return arr[n]
    else:
        result = fib(n-1) + fib(n-2)
        arr[n] = result
        return result
```
