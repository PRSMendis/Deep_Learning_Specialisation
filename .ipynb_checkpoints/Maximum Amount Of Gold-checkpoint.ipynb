```python
def maximum_gold(capacity, weights):
#     A = [num for num in range(0,capacity+1)]
    A = np.zeros((len(weights) + 1, capacity+1), dtype=int)
    for i, weight in enumerate(weights):
        for j in range(weight,capacity+1):
            A[i, j] = weight
    return A
maximum_gold(10, (3, 5, 3, 3, 5))
```

```python
def maximum_gold(capacity, weights):
#     A = [num for num in range(0,capacity+1)]
    A = np.zeros((len(weights), capacity+1), dtype=int)
    for i, weight in enumerate(weights):
        for j in range(weight,capacity+1):
            A[i, j] = max(weight, A[i-1,j], (weight + A[i-1,j-weight]))
    return A[-1,-1]
maximum_gold(10, (3, 5, 3, 3, 5))
```

```python
def maximum_gold(capacity, weights):
#     A = [num for num in range(0,capacity+1)]
    A = np.zeros((len(weights), capacity+1), dtype=int)
    for i, weight in enumerate(weights):
        for j in range(1,capacity+1):
            if weight > i:
                A[i,j] = A[i-1,j]
                A[i, j] = max(weight, A[i-1,j], (weight + A[i-1,j-weight]))
            else:
                A[i,j] = A[i-1,j]
                A[i, j] = max(weight, A[i-1,j], (weight + A[i-1,j-weight]))
    return A
maximum_gold(10, (1, 4, 8))
```
