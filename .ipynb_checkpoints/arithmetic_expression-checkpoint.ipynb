```python
import math
import operator

def calc(a, b, op):
    print(a,b)
    ops = {
        '+' : operator.add,
        '-' : operator.sub,
        '*' : operator.mul,
        '/' : operator.truediv,  # use operator.div for Python 2
        '%' : operator.mod,
        '^' : operator.xor,
    }
    
    ops[op](a, b) 

#     if op == '+':
#         return a + b
#     elif op == '-':
#         return a - b
#     else:
#         return a * b


def MinAndMax(M, m, i, j, operators):

    min_value = math.inf
    max_value = -math.inf
    for k in range(i, j):
        a = calc(M[i][k], M[k+1][j], operators[k])
        b = calc(M[i][k], m[k+1][j], operators[k])
        c = calc(m[i][k], M[k+1][j], operators[k])
        d = calc(m[i][k], m[k+1][j], operators[k])
        min_value = min(min_value, a, b, c, d)
        max_value = max(max_value, a, b, c, d)
    return min_value, max_value


def get_maximum_value(dataset):
    
    operators, operands = [], []

    for i in dataset:
        if i in ['+', '-', '*']:
            operators.append(i)
        else:
            operands.append(int(i))
    

    
    operations = [num for num in dataset if not num.isnumeric()]
    operands = operations

    n = len(operands)
    m = [[None for x in range(n)] for x in range(n)]
    M = [[None for x in range(n)] for x in range(n)]

    for i in range(n):
        m[i][i] = operands[i]
        M[i][i] = operands[i]

    for s in range(1, n):
        for i in range(0, n-s):
            j = i + s
            m[i][j], M[i][j] = MinAndMax(M, m, i, j, operators)

    return M[0][n-1]
```

```python
import numpy as np
import operator

ops = {
    '+' : operator.add,
    '-' : operator.sub,
    '*' : operator.mul,
}

def find_maximum_value(dataset):
    numbers = [num for num in dataset if num.isnumeric()]
    x = y = len(numbers)
    operations = [num for num in dataset if not num.isnumeric()]
    A = np.zeros((x,y, 2), dtype=int)
    for i in range(1,x):
        for j in range(1,y):
            A[i,j] = i+j
    for i, num in enumerate(numbers):
        A[i,i] = num
    print(A[:,:,0])
    i = 0
    j = 1
    k = 0
    q = 1
    
    d = 1
    
    while d < y:
        j = d
        
        while j < y :
            
            print('yaaa')
            A[i,j, k] = q
            q+=1
#             print(f"A:{type(A)}, cart:{type((i,j))}")
#             A[i,j, k] = max_at_pos(A=A, cart=(i,j), op=ops[operations[i]])
#             print(max_at_pos(A=A, cart=(i,j), op=ops[operations[i]]))
            print(f"D:{d}")
            A[i,j, k] = max(max_at_pos(A=A, cart=(i,j), op=ops[operations[i]]))
#             A[i,j, 0] = max(ops[operations[i]](A[i,j-1,0], A[i+1,j,0]), ops[operations[i]](A[i,j-1,1], A[i+1,j,1]))
#             A[i,j, 1] = min(ops[operations[i]](A[i,j-1,0], A[i+1,j,0]), ops[operations[i]](A[i,j-1,1], A[i+1,j,1]))
            
    
            i+=1
            j+=1
        i = 0
        d += 1
        j = d
        
            
    

    return A[:,:,0]


def max_at_pos(A, cart, op):
    '''
    Input: Array, column, cartesion position
    Output: largest possible value (negative, and pos)
    
    Number of times needed to iterate = j-i
    '''
    i, j = cart
    Results = []
    k = 0
    print(A[:,:,0])
    for z in range(0,j-i):
#         x = A[i+1+z,j,k] + A[i,z,k]
#         x = A[i+1+z,j,k] + A[i,z,k]
        print(f"op:{op}, A[i,(0+z),k]:{A[i,(i+(1+z)),k]}, A[i+1+z,j,k]:{A[i+1+z,j,k]}")
        print(f" i:{i}, j:{j}, z:{z}, k:{k}")
        x = op(A[i,i+(z),k], A[i+1+z,j,k])
    
        Results.append(x)
    print(Results)
    return Results

B = find_maximum_value(i[0])
print("-------")
print(B)
```
